```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/Dataset of Diabetes .csv')
data.head()
```

```python
data.info()
```

```python
data = data.drop(['ID'],axis=1)
data.info()
```

```python
df = data.drop(['No_Pation'],axis=1)
df.info()
```

```python
df['CLASS'] = df['CLASS'].replace({'N ':'N','Y ':'Y'})
df['CLASS'].unique()
```

```python
df['Gender'].unique()
```

```python
df['Gender'] = df['Gender'].replace({'f':'F'})
df['Gender'].unique()
```

```python
classes = df.groupby(['CLASS'])
n = classes.get_group('N')
p = classes.get_group('P')
y = classes.get_group('Y')
```

```python
sns.histplot(n['Gender'])
```

```python
sns.histplot(p['Gender'])
```

```python
sns.histplot(y['Gender'])
```

```python
n_men = df[df['Gender']=='M']['Gender'].count()
n_fem = df[df['Gender']=='F']['Gender'].count()
```

```python
print("non-diabetic men: ",(n[n['Gender']=='M']['Gender'].count()/n_men)*100)
print("non-diabetic women: ",(n[n['Gender']=='F']['Gender'].count()/n_fem)*100)
print("pre-diabetic men: ",(p[p['Gender']=='M']['Gender'].count()/n_men)*100)
print("pre-diabetic women: ",(p[p['Gender']=='F']['Gender'].count()/n_fem)*100)
print("diabetic men: ",(y[y['Gender']=='M']['Gender'].count()/n_men)*100)
print("diabetic women: ",(y[y['Gender']=='F']['Gender'].count()/n_fem)*100)
```

Men have a higher chance of acquiring diabetes than woman

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
x = pd.get_dummies(df.drop(['CLASS'],axis=1),drop_first=True)
y = df['CLASS']
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
```

```python
model1 = DecisionTreeClassifier(random_state=0)
model1.fit(x_train,y_train)
```

```python
y_pred = model1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,balanced_accuracy_score
```

```python
print("Score: ",accuracy_score(y_test,y_pred))
print("Score: ",balanced_accuracy_score(y_test,y_pred))
```
